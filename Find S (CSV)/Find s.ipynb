```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

df = pd.read_csv('../input/inputcsv/input.csv', header=None)
print(df)
train = df.values

hyp = ["#" for _ in range(len(train[0])-1)]
for i in range(len(train)):
        if train[i][-1] == "Yes":
            for j in range(len(train[0])-1):
                if hyp[j] != train[i][j] and hyp[j] == "#":
                    hyp[j] = train[i][j]
                elif hyp[j] != train[i][j] and hyp[j] != "#":
                    hyp[j] = "?"

                    
print("Hypothesis:-",hyp)
```
