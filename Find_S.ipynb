```python
r = int(input("Enter number of rows:"))
c = int(input("Enter number of col:"))

matrix = []

for i in range(r):
    a=[]
    for j in range(c):
        a.append(input())
    matrix.append(a)
    
hyp = ["#"]*c

for j in range(c):
    hyp[j] = matrix[0][j]

for i in range (r):
    for j in range(c):
        if(matrix[i][c-1]=="T"):
            if(matrix[i][j]!=hyp[j]):
                hyp[j]="?"
            else:
                hyp[j]=matrix[i][j]
                
print(hyp[:c-1])                
```
